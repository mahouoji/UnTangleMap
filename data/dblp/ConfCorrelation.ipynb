```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df = pd.read_csv('./data/authorconf.csv').set_index('author')
df
```

```python
plot.subplots(figsize=(20,15))
sns.heatmap(df)
```

```python
del df['ITCS']
```

```python
corr_spearman = df.corr('spearman')
plot.subplots(figsize=(20,15))
sns.heatmap(corr_spearman, 
            xticklabels=corr_spearman.columns,
            yticklabels=corr_spearman.columns,
            cmap='RdBu_r',
            annot=False,
            center=0,
            linewidth=0.1)
```

```python
corr_spearman.to_csv('./data/corr_spearman.csv')
```
