```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
```

```python
df = pd.read_csv('./data/author_conf.csv').set_index('author')
df
```

```python
plot.subplots(figsize=(20,15))
sns.heatmap(df)
```

```python
del df['ITCS']
```

```python
corr_spearman = df.corr('spearman')
plot.subplots(figsize=(20,15))
sns.heatmap(corr_spearman, 
            xticklabels=corr_spearman.columns,
            yticklabels=corr_spearman.columns,
            cmap='RdBu_r',
            annot=False,
            center=0,
            linewidth=0.1)
corr_spearman.to_csv('./data/corr_spearman.csv')
```

```python
corr_kendall = df.corr('kendall')
plot.subplots(figsize=(20,15))
sns.heatmap(corr_kendall, 
            xticklabels=corr_kendall.columns,
            yticklabels=corr_kendall.columns,
            cmap='RdBu_r',
            annot=False,
            center=0,
            linewidth=0.1)
corr_kendall.to_csv('./data/corr_kendall.csv')
```

```python
corr_pearson = df.corr('pearson')
plot.subplots(figsize=(20,15))
sns.heatmap(corr_pearson, 
            xticklabels=corr_pearson,
            yticklabels=corr_pearson,
            cmap='RdBu_r',
            annot=False,
            center=0,
            linewidth=0.5)
corr_pearson.to_csv('./data/corr_pearson.csv')
```
